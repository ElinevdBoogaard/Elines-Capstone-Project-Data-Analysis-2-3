Connect to the database.

Import pandas module.


```python
import pandas as pd
```

## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by Age
order by Age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
query="""SELECT * FROM master"""
df_master=pd.read_sql_query(query, conn)
```

```python
df_master[["ConvertedComp"]].hist(bins=8)
```

### Box Plots


Plot a box plot of `Age.`


```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_master["Age"].plot(kind='box')
plt.title('Boxplot of Age')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
df_master.plot(kind='scatter', x="Age", y="WorkWeekHrs")
plt.title('Scatterplot of the relation between Workweek Hours and Age')
plt.xlabel('Age')
plt.ylabel('Number of Hours in a Workweek')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
norm_age=(df_master["Age"]-df_master["Age"].min())/(df_master["Age"].max()-df_master["Age"].min())
df_master.plot(kind='scatter', x="CodeRevHrs", y="WorkWeekHrs", s=norm_age*1000, alpha=0.5)
plt.title('Bubbleplot of the relation between Workweek Hours, CodeRevHrs and Age')
plt.xlabel('CodeRevHrs')
plt.ylabel('Number of Hours in a Workweek')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
table_name = 'DatabaseDesireNextYear'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

```python
QUERY = """
SELECT DatabaseDesireNextYear, Count(*) as Count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

df_DDNY = pd.read_sql_query(QUERY, conn)
df_DDNY.set_index('DatabaseDesireNextYear', inplace=True)
labels = df_DDNY.index
plt.figure(figsize=(8, 6))
plt.pie(df_DDNY['Count'], labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.8)
plt.title('Top 5 Database Desire Next Year')
plt.axis('equal')
plt.show()
```

```python
QUERY = """
SELECT LanguageDesireNextYear, Count(*) as Count
FROM LanguageDesireNextYear
GROUP BY LanguageDesireNextYear
ORDER BY Count DESC
LIMIT 5
"""

df_LDNY = pd.read_sql_query(QUERY, conn)
df_LDNY.set_index('LanguageDesireNextYear', inplace=True)
labels = df_LDNY.index
plt.figure(figsize=(8, 6))
plt.pie(df_LDNY['Count'], labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.8)
plt.title('Top 5 Languages Desired Next Year')
plt.axis('equal')
plt.show()
```

```python
#hoeveel respondenten werken nu met taal SQL?
QUERY = """
SELECT LanguageWorkedWith, Count(*) as Count
FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY Count DESC
LIMIT 5
"""
df_LWW = pd.read_sql_query(QUERY, conn)
df_LWW.head()
```

```python
#hoeveel respondenten werken nu met database mySQL?
QUERY = """
SELECT DatabaseWorkedWith, Count(*) as Count
FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
ORDER BY Count DESC
LIMIT 5
"""
df_DWW = pd.read_sql_query(QUERY, conn)
df_DWW.head()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
df_master_30to35= df_master[(df_master['Age']>=30)&(df_master['Age']<35)]

order=["WorkWeekHrs", "CodeRevHrs"]
plt.figure(figsize=(10, 6))
df_master_30to35.groupby('Age')[order].median().plot.bar(stacked=True)
plt.title('Median WorkWeekHrs and CodeRevHrs for the age group 30 to 35')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
df_master_40to65= df_master[(df_master['Age']>=45)&(df_master['Age']<60)]

plt.figure(figsize=(8, 5))
df_master_40to65.groupby('Age')["ConvertedComp"].median().plot(kind='line')
plt.title('Median ConvertedComp for ages from 45 to 60')
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
query="""
SELECT Mainbranch, Count(*) as Count
FROM master
GROUP BY Mainbranch
"""
df_MB=pd.read_sql_query(query, conn)
df_MB.set_index('MainBranch', inplace=True)
df_MB.plot(kind='barh')
plt.title('MainBranch')
plt.show()
```
